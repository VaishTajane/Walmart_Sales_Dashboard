```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Excel file
df = pd.read_excel("Walmart.xlsx", sheet_name="Walmart")
```

```python
# Display basic information
df.info()
print(df.head())
```

```python
# Checking for missing values
print("Missing Values:\n", df.isnull().sum())
```

```python
# Basic statistics
print(df.describe())
```

```python
# Sales Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Sales'], bins=50, kde=True)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Profit vs Sales
plt.figure(figsize=(10,5))
sns.scatterplot(x=df['Sales'], y=df['Profit'])
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
# Category-wise Sales
plt.figure(figsize=(12,6))
sns.barplot(x=df['Category'].value_counts().index, y=df['Category'].value_counts().values)
plt.xticks(rotation=45)
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Number of Sales')
plt.show()
```

```python
# State-wise Sales
plt.figure(figsize=(15,6))
df.groupby('State')['Sales'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Total Sales by State')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.show()
```
